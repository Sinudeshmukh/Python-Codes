```python
# importing pandas as pd
import pandas as pd

df = pd.read_csv("nba.csv")
df
```

```python
df.sort_values(by=['Weight','Height'],ascending=False)
```

```python
grpbyteam = df.groupby('Team')
grpbyteam.first()  #list 
```

```python
grpbyteam.ngroups  #all grps count
```

```python
grpbyteam.groups  # all grps records
```

```python
grpbyteam['Age'].max()
```

```python
grpbyteam['Age'].agg(['min','max','count','median','mean'])
```

```python
grpbyteam_posn = df.groupby(['Team','Position'])
# pd.set_option('display.max_rows',300)
# grpbyteam_posn.first()
grpbyteam_posn.sort_values()
```

```python
boston = grpbyteam.get_group('Boston Celtics')
```

```python
boston.sort_values(['Weight'],ascending=False)
```

```python
grpbyteam.get_group('Boston Celtics').Weight.max()
```
